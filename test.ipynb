```python
from loan_calculator import  loan_simulator
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
result = loan_simulator(
    loan_amount=1800000, annual_interest_rate=11, emi=20000, extra_payment=20000, extra_strategy="alternate")

df = result["schedule"]
print(result["months"], result["years"], result["total_interest"])

```

```python
# Without extra payment
base = loan_simulator(
    loan_amount=1800000,
    annual_interest_rate=11,
    emi=20000,
    extra_payment=0
)

df_base = base["schedule"]

plt.figure()

plt.plot(df_base["Month"], df_base["Closing Balance"], label="Normal EMI")
plt.plot(df["Month"], df["Closing Balance"], label="With Extra Payment")

plt.xlabel("Month")
plt.ylabel("Loan Balance")
plt.title("Impact of Extra Payments")

plt.legend()

plt.show()
```

```python
plt.figure()

plt.plot(df["Month"], df["Closing Balance"])

plt.xlabel("Month")
plt.ylabel("Loan Balance")
plt.title("Loan Balance Over Time")

plt.show()
```

```python
plt.figure()

plt.plot(df["Month"], df["Interest"], label="Interest")
plt.plot(df["Month"], df["Principal Paid"], label="Principal")

plt.xlabel("Month")
plt.ylabel("Amount")
plt.title("Interest vs Principal Over Time")

plt.legend()

plt.show()
```
